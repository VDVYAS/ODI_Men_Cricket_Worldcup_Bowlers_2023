# Analysis

### Wickets

```python
df.columns
```

```python
wickets= df.groupby(['player'], as_index=False)['wickets'].sum().sort_values(by='wickets',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=wickets,x='player',y='wickets')

for bars in ax.containers:
    ax.bar_label(bars)
```

 above graph we can see mohammed shami is highest wicker taker in ODI mens cricket world cup in 2023

### maidens

```python
maidens= df.groupby(['player'], as_index=False)['maidens'].sum().sort_values(by='maidens',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=maidens,x='player',y='maidens')

for bars in ax.containers:
    ax.bar_label(bars)
```

from above we can see japrit bumrah is highest maidens over 

### runs

```python
runs= df.groupby(['player'], as_index=False)['runs'].sum().sort_values(by='runs',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=runs,x='player',y='runs')

for bars in ax.containers:
    ax.bar_label(bars)
```

in above graph we can see harish rauf given highest run in this world cup

```python
os= df.groupby(['player'], as_index=False)['0s',].sum().sort_values(by='0s',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=os,x='player',y='0s')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
four_s= df.groupby(['player'], as_index=False)['4s',].sum().sort_values(by='4s',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=four_s,x='player',y='4s')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
six_s= df.groupby(['player'], as_index=False)['6s',].sum().sort_values(by='6s',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=six_s,x='player',y='6s')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
team_runs= df.groupby(['team'], as_index=False)['runs'].sum().sort_values(by='runs',ascending=False).head()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=team_runs,x='team',y='runs')

for bars in ax.containers:
    ax.bar_label(bars)
```

In above graph we can say that australian bowlers give highest run 

```python
team_runs_low= df.groupby(['team'], as_index=False)['runs'].sum().sort_values(by='runs',ascending=False).tail()
sns.set(rc={'figure.figsize' : (9,5)})
ax=sns.barplot(data=team_runs_low,x='team',y='runs')

for bars in ax.containers:
    ax.bar_label(bars)
```

in above graphh we can say that afghanistan bowlers given lowest run

```python
highest_wicket_team=df.groupby(['team'], as_index=False)['wickets'].sum().sort_values(by='wickets',ascending=False).head()
sns.set(rc={"figure.figsize" : (10,5)})
ax=sns.barplot(data=highest_wicket_team,x='team',y='wickets')
for bars in ax.containers:
    ax.bar_label(bars)
```

in above graph india is taking hightest wicket in this world cup

```python
lowest_team_wickets=df.groupby(['team'],as_index=False)['wickets'].sum().sort_values(by='wickets',ascending=False).tail()
sns.set(rc={"figure.figsize" : (10,5)})
ax=sns.barplot(data=lowest_team_wickets,x='team',y='wickets')
for bars in ax.containers:
    ax.bar_label(bars)
```

in above graph we can say that sri lanka is taken lowest weicket in this world cup
